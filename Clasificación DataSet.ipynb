# Funcion para asignar categoria

```python
def asignaCat(text):
    tmp = str(text)
    if any(word in tmp for word in transporte):
        return 'transporte'
    elif any(word in tmp for word in bien):
        return 'bien'
    elif any(word in tmp for word in electronico):
        return 'electronico'
    elif any(word in tmp for word in herramientas):
        return 'herramientas'
    elif any(word in tmp for word in accesorios):
        return 'accesorios'
    elif any(word in tmp for word in pinturas):
        return 'pinturas'
    else:
        return asignaSubCat(tmp)
    
```

# Funcion para asignar subcategoria

```python
def asignaSubCat(text):
    tmp = str(text)
    if any(word in tmp for word in terrestre):
        return 'terrestre'
    elif any(word in tmp for word in aereo):
        return 'aereo'
    elif any(word in tmp for word in marino):
        return 'marino'
    elif any(word in tmp for word in inmobiliario):
        return 'inmueble'
    elif any(word in tmp for word in mobiliario):
        return 'mueble'
    elif any(word in tmp for word in telecomunicaciones):
        return 'telecomunicaciones'
    elif any(word in tmp for word in cocina):
        return 'cocina'
    elif any(word in tmp for word in pinturas):
        return 'pinturas'
    elif any(word in tmp for word in emergencias):
        return 'emergencias'
    elif any(word in tmp for word in electricos):
        return 'electricos'
    elif any(word in tmp for word in jardineria):
        return 'jardineria'
    elif any(word in tmp for word in trabajo):
        return 'trabajo'
    elif any(word in tmp for word in hogar):
        return 'hogar'
    elif any(word in tmp for word in papeleria):
        return 'papeleria'
    elif any(word in tmp for word in dispositivos):
        return 'dispositivos'
    elif any(word in tmp for word in arte):
        return 'arte'
    else:
        return 'NO CLASIFICADO'
        
```

### Funcion para tratar datos

```python
def limpiar(text):
    #convertir a minusculas
    text = text.lower()
    #eliminar liks
    text = re.sub('http\S+', ' ', text)
    #eliminar expresiones regulares
    text = re.sub(regex , ' ', text)
    #eliminar numeros
    text = re.sub("\d+", ' ', text)
    #eliminar multiples espacios en blanco
    text = re.sub("\\s+", ' ', text)
    text_tokens = word_tokenize(text)
    #print(text_tokens)

    tokens_without_sw = [word for word in text_tokens if not word in others and len(word) > 1]
    return (tokens_without_sw)
#limpiar()
```
